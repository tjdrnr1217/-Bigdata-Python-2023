### 주피터 노트북 디버깅

````python
# 나누기 함수
def divide(x, y):
    result = ''
    try:
   
        result = x // y
    except Exception as e:
        print(f'예외발생 : {e}')
    
    print('나누기 완료')
    return result

print(divide(7, 0))
````
