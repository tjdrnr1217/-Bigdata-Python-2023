##### 결측치 확인, 제거, 채우기
- 데이터분석 분석이전에 하는 일중 제일 중요한 일 


```python
datas = {
    '이름' : ['길동이','길순이','유고','애슐리'],
    '국어' : [90,100,None,99],
    '영어' : [None,100,99,100],
    '수학' : [50,100,None,20]
    }
df5=pd.DataFrame(datas)
df5
```

```python
df5.isnull()
```

```python
df5.isnull().sum()
```

```python
# 샤용빈도수 거의제로
df5.notnull().sum()
```

```python
# 결측치 채우기
df5['국어'] = df5['국어'].fillna(50)
df5
```

```python
df5['영어'].fillna(40, inplace=True)
df5
```

```python
df5['수학'].fillna(method='ffill') # foreward 값으로 채워넣어라
```

```python
df5['수학'].fillna(method='bfill', inplace=True) #backward 값으로 채워넣어라
```

```python
#결측치 삭제
df7 =pd.DataFrame(datas)
df7
```

```python
# 형식변환
df7['국어']
```

```python
df7['국어'].astype('object')
```

```python
# 추가 NaN이 있으면 float64 -> Int64로 변경하려면 결측치가 없어야 변경가능
df7['국어'].fillna(30).astype('int64')
```

```python
df7
```

```python
for index, row in df7.iterrows():
    print(index)
    print(row)
    print('-'* 20) # 가로줄 만들어라
```

```python
for row in df7.itertuples():
    print(row)
```
