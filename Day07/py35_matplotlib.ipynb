##### matplotilib 사용

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
arr = np.random.randn(30)
arr
````

````python
plt.plot(arr)
````

````python
# 다중차트
fig = plt.figure() # Figure 객체 생성
ax1 = fig.add_subplot(2,2,1) # 차트1(2행 2열 차트중 1)
ax2 = fig.add_subplot(2,2,2) # 차트2(2행 2열 차트중 2)
ax3 = fig.add_subplot(2,2,3) # 차트3(2행 2열 차트중 3)

ax1.hist(np.random.randn(100)) # 분포표
ax2.scatter(np.arange(30), np.random.randn(30)) # 스케터차트
ax3.plot(np.random.randn(50).cumsum()) # 라인차트

plt.show() # 차트보기
````

````python
# 안쓰는 차트도 같이 나옴
fig, axes = plt.subplots(2,2)
axes[0][0].hist(np.random.randn(100)) # 분포표
axes[0][1].scatter(np.arange(30), np.random.randn(30))
axes[1][1].plot(np.random.randn(50).cumsum())

plt.show()
````

##### 차트꾸미기

````python
# 차트색상, 라인종류, 마커
fig, axes = plt.subplots(2,2)
axes[0][0].hist(np.random.randn(100), bins=20, color='green', alpha=0.5) # bins = 범위 20개 늘림, alpha = 투명도(0.0[투명도] ~1.0[불투명])
axes[0][1].scatter(np.arange(30), np.random.randn(30), color='lightgreen')
axes[1][0].plot(np.random.randn(50).cumsum(), linestyle=':', color='red', marker='1') # marker = 값위치표시 *(별), ^(삼각형)
axes[1][1].plot(np.random.randn(50).cumsum(), linestyle='--', marker='o') # linestyle = 선종류 -(일산선) --(대시), :(점선), -.(선점)

plt.show()
````

````python
# 범례 legend
plt.plot(np.random.randn(50).cumsum(), label='x')
plt.plot(np.random.randn(50).cumsum(), label='x')
plt.legend(loc='right')# 'best' 'upper left', 'upper right', 'lower left', 'lower right', 'right

plt.show()
````

##### Pandas 데이터프레임, 시리즈

````python
#판다스 시리즈
pd.Series(np.random.randn(1000).cumsum()).plot()
plt.show()
````

````python
# 데이터프레임 plot
pd.DataFrame(np.random.randn(10,4).cumsum(0), columns=['A', 'B', 'C', 'D'], index=np.arange(0,100,10)).plot()
````

````python
# x, y출 설정
chart = pd.Series(np.random.randn(1000).cumsum()).plot()
chart.set_xticks([0,250,500,750,1000]) # x축 간격 조정
chart.set_yticks([0,25, 50]) # y축 간격 조정
chart.set_title('Random Linechart') # 영어는 잘되는데 한글은 기본 깨짐
chart.set_xticklabels(['zero', 'one', 'two', 'thress', 'four'], rotation=300) # x축을 라벨로 표시

plt.show() #쓸데없는 타입, 결과 메시지 삭제
````

````python
# 한글 깨짐현상
import matplotlib
from matplotlib import font_manager,rc

font_path ='C:/Windows/Fonts/HanSantteutDotum-Bold.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font)
matplotlib.rcParams['axes.unicode_minus'] = False

chart = pd.Series(np.random.randn(1000).cumsum()).plot()
chart.set_xticks([0,250,500,750,1000]) # x축 간격 조정
chart.set_yticks([0,25, 50]) # y축 간격 조정
chart.set_title('랜덤 Linechart') # 영어는 잘되는데 한글은 기본 깨짐
chart.set_xticklabels(['zero', 'one', 'two', 'thress', 'four'], rotation=300) # x축을 라벨로 표시
chart.grid(True) # 그리드선 표시

plt.show() #쓸데없는 타입, 결과 메시지 삭제
````

##### 예제 연습

캐글 타이타닉 연습문제 csv 

https://raw.githubusercontent.com/developer-sdk/kaggle-python-beginner/master/datas/kaggle-titanic/train.csv

````python
train = pd.read_csv('https://raw.githubusercontent.com/developer-sdk/kaggle-python-beginner/master/datas/kaggle-titanic/train.csv')

train.tail()
````

````python
# 성별 생존률교차초표 생성
gender_Survived = pd.crosstab(train['Sex'], train['Survived'])
gender_Survived.columns = gender_Survived.columns.map({0:"Dead", 1:"Alive"})
````

````python
# bar = gender_Survived.plot.bar() #바차트
bar = gender_Survived.plot(kind='bar', title='생존자/사망자')

# 차트의 수치표시
for p in bar.patches:
    left, bottom, width, height = p.get_bbox().bounds
    bar.annotate(f'{height:n}',(left+width/2, height*1.01), ha='center')

# gender_Survived.plot.bar() # 바차트
gender_Survived.plot.bar(subplots=True)
gender_Survived.plot.bar(x='Dead')
# bar.set_title('생존자/사망자수')
# gender_Survived.plot.bar(x='Alive')
````

````python
# 가로 바차트
gender_Survived.plot(kind='barh')
gender_Survived.plot.barh(subplots=True)
````

````python
# 라인차트
gender_Survived.plot(kind='line')
````

````python
#파이차트
gender_Survived['Alive'].plot(kind='pie', title='생존자 비율', autopct='%1.1f%%')
````
